```python
import pandas as pd
import numpy as np
from CreateModel import GenerateUserMovieRating
from sklearn.model_selection import train_test_split

movie_ratings = GenerateUserMovieRating()
```

```python
movie_ratings_train, movie_ratings_test = train_test_split(movie_ratings, test_size=0.2, random_state=10)
```

```python
# make a copy of the original df so we can zero some of the ratings
movie_ratings_zeroed = movie_ratings.copy()

# zero-out half of the ratings of each user in the test set
for userId in movie_ratings_test.index.tolist():
    # calculate how many ratings from this user should be deleted (half of the number of ratings this user did)
    numRatingsToDelete = movie_ratings_test.loc[userId].ne(0).sum()//2
    
    # get this user's row
    user_row = movie_ratings_test.loc[userId]

    # make ratings zero until we've met the count that we need to delete
    countRatings = 0
    for movie_name, rating in user_row.items():
        if rating > 0 and countRatings < numRatingsToDelete:
            countRatings +=1
            movie_ratings_zeroed.loc[userId, movie_name] = 0
        if countRatings == numRatingsToDelete:
            break
```

```python
from CreateModel import matrix_factorization

# calculate matrix factorization with the zeroed-out ratings
movie_titles = movie_ratings_zeroed.columns.to_list()
R = np.array(movie_ratings_zeroed)
N = len(R)
M = len(R[0])
K = 2
kList = [10]
for testK in kList: # test
    K = testK
    P = np.random.rand(N,K)
    Q = np.random.rand(M,K)
    # Steps = 5000
    Steps = 500
    nP, nQ = matrix_factorization(R, P, Q, K, steps=Steps)
    nR = np.dot(nP, nQ.T)
    nR_df = pd.DataFrame(nR, columns=movie_titles)
    print(nR_df.head())

    nR_df.to_csv("RatingsModelFromZeroed2_k_{}_steps_{}.csv".format(K,Steps), index=False)
```

```python
user_id_mapping = pd.Series(movie_ratings_zeroed.index.to_list())
```

```python
# map new column called userId
nR_df['userId'] = nR_df.index.map(user_id_mapping)

# Get the index of the "userID" column
user_id_index = nR_df.columns.get_loc("userId")

# Insert the "userID" column as the first column
nR_df.insert(0, "userId", nR_df.pop("userId"))

nR_df
```

```python
nR_df.set_index("userId", inplace=True)
nR_df
```

```python
# calculate MSE based on actual ratings that got zeroed out earlier

count = 0
square_sum = 0
for userId in movie_ratings_test.index.tolist():
    # get this user's row
    user_row = movie_ratings_test.loc[userId]

    # check each rating associated with this user
    for movie_name, rating in user_row.items():
        # if the rating is zero in the zeroed matrix but not zero in the original ratings matrix, then add this item to
        # the MSE by calculating error between the original ratings matrix and the model ratings 
        if movie_ratings_zeroed.loc[userId, movie_name] == 0 and movie_ratings.loc[userId, movie_name] != 0:
            count +=1
            difference = movie_ratings.loc[userId, movie_name] - nR_df.loc[userId, movie_name]
            square_sum += (difference) ** 2

mse = square_sum / count
print(mse)
```
